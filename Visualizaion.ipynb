```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
weight_decay = False

if weight_decay == True: 
    k = 4
    leng = 50
else: 
    k = 6
    leng = 100

loss_train = np.zeros((k,leng))
acc_train = np.zeros((k,leng))
loss_val = np.zeros((k,leng))
acc_val = np.zeros((k,leng))

if weight_decay == True:
    t = ['C_R20_lr01', 'C_R20_lr005', 'V_R20_lr01', 'V_R20_lr005']
    for i in range(k):
        acc_train[i] = np.load('npy_dep/'+t[i] +'_E50_train_acc.npy')
        loss_train[i] = np.load('npy_dep/'+t[i] +'_E50_train_loss.npy')
        acc_val[i] = np.load('npy_dep/'+t[i] +'_E50_val_acc.npy')
        loss_val[i] = np.load('npy_dep/'+t[i] +'_E50_val_loss.npy')
else:
    t = ['C_R20_lr01', 'C_R20_lr005', 'C_R20_lr001', 'V_R20_lr01', 'V_R20_lr005', 'V_R20_lr001']
    for i in range(k):
        acc_train[i] = np.load('npy_dep/'+t[i] +'_E100_train_acc.npy')
        loss_train[i] = np.load('npy_dep/'+t[i] +'_E100_train_loss.npy')
        acc_val[i] = np.load('npy_dep/'+t[i] +'_E100_val_acc.npy')
        loss_val[i] = np.load('npy_dep/'+t[i] +'_E100_val_loss.npy')
```

```python
plt.figure(figsize = (12,9))
x = np.arange(leng)

if weight_decay == True:
    plt.plot(x, acc_train[0], label = 'CBAM, lr= 0.1', c = 'red')
    plt.plot(x, acc_train[1], label = 'CBAM, lr= 0.05', c = 'orange')
    plt.plot(x, acc_train[2], label = 'Vanilla, lr = 0.1', c = 'darkblue')
    plt.plot(x, acc_train[3], label = 'Vanilla, lr= 0.05', c = 'blue')
else:
    plt.plot(x, acc_train[0], label = 'CBAM, lr= 0.1', c = 'red')
    plt.plot(x, acc_train[1], label = 'CBAM, lr= 0.05', c = 'orange')
    plt.plot(x, acc_train[2], label = 'CBAM, lr= 0.01', c = 'pink')
    plt.plot(x, acc_train[3], label = 'Vanilla, lr = 0.1', c = 'darkblue')
    plt.plot(x, acc_train[4], label = 'Vanilla, lr= 0.05', c = 'blue')
    plt.plot(x, acc_train[5], label = 'Vanilla, lr = 0.01', c = 'lightblue')

plt.title('Train accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylim([30,105])
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (12,9))


if weight_decay == True:
    plt.plot(x, acc_val[0], label = 'CBAM, lr= 0.1', c = 'red')
    plt.plot(x, acc_val[1], label = 'CBAM, lr= 0.05', c = 'orange')
    plt.plot(x, acc_val[2], label = 'Vanilla, lr = 0.1', c = 'darkblue')
    plt.plot(x, acc_val[3], label = 'Vanilla, lr= 0.05', c = 'blue')
else:
    plt.plot(x, acc_val[0], label = 'CBAM, lr= 0.1', c = 'red')
    plt.plot(x, acc_val[1], label = 'CBAM, lr= 0.05', c = 'orange')
    plt.plot(x, acc_val[2], label = 'CBAM, lr= 0.01', c = 'pink')
    plt.plot(x, acc_val[3], label = 'Vanilla, lr = 0.1', c = 'darkblue')
    plt.plot(x, acc_val[4], label = 'Vanilla, lr= 0.05', c = 'blue')
    plt.plot(x, acc_val[5], label = 'Vanilla, lr = 0.01', c = 'lightblue')

plt.title('Validation accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylim([30,105])
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
